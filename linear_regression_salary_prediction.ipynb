# Simple Linear Regression - Salary Prediction

## Problem
Predict salary based on experience

## Objective
Understand relationship between experience and salary

1. Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

2. Create Dataset

```python
data = {
    "Experience": [1, 2, 3, 4, 5],
    "Salary": [30000, 40000, 50000, 60000, 70000]
}

df = pd.DataFrame(data)
df
```

3. Define X and y

```python
X = df[["Experience"]]
y = df["Salary"]
```

4. Train Model

```python
model = LinearRegression()
model.fit(X, y)
```

5. Prediction

```python
predictions = model.predict(X)
print(predictions)
```

6. Visualization

```python
plt.scatter(X, y)
plt.plot(X, predictions)
plt.title("Experience vs Salary")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.grid()
plt.show()
```

```python
exp = [[5]]
pred_salary = model.predict(exp)
print("Predicted salary for 5 years:", pred_salary[0])
```

## Conclusion

- Salary increases as experience increases  
- The model shows a clear linear relationship  
- Predicted salary for 6 years of experience gives a practical use case
