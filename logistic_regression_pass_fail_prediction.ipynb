# Logistic Regression - Pass/Fail Prediction

## Objective
To predict whether a student will pass or fail based on study hours

## Problem Statement

Predict if a student will pass (1) or fail (0) using study hours.
This is a classification problem.

1. Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

2. Create Dataset

```python
data = {
    "Hours": [1, 2, 3, 4, 5, 6, 7, 8],
    "Result": [0, 0, 0, 0, 1, 1, 1, 1]
}

df = pd.DataFrame(data)
df
```

3. Define Features and Target

```python
X = df[["Hours"]]
y = df["Result"]
```

4. Train Logistic Regression Model

```python
model = LogisticRegression()
model.fit(X, y)
```

5. Predictions

```python
predictions = model.predict(X)
print(predictions)
```

6. Custom Prediction

```python
hours = [[5]]
pred = model.predict(hours)
print("Prediction (0=Fail, 1=Pass):", pred[0])
```

7. Prediction Probability

```python
prob = model.predict_proba(hours)
print("Probability [Fail, Pass]:", prob)
```

8. Visualization

```python
plt.figure(figsize=(6,4))

plt.scatter(X, y, s=100)
plt.plot(X, model.predict_proba(X)[:,1])

plt.yticks([0, 1], ["Fail", "Pass"])

plt.title("Logistic Regression (Pass/Fail Prediction)")
plt.xlabel("Study Hours")
plt.ylabel("Result")
plt.grid()

plt.show()
```

## Conclusion

- Logistic regression is used for classification problems  
- It predicts outcomes like pass (1) or fail (0)  
- It provides probability using a sigmoid curve
