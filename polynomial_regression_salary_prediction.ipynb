# Polynomial Regression - Salary Prediction

## Objective
To understand non-linear relationship between experience and salary

## Problem Statement

We want to predict salary based on experience where the relationship is not linear.
Polynomial regression helps capture curved patterns in data.

1.Code (Import)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

2.Code (Dataset)

```python
data = {
    "Experience": [1, 2, 3, 4, 5],
    "Salary": [30000, 45000, 70000, 100000, 150000]
}

df = pd.DataFrame(data)
df
```

3.Code (X & y)

```python
X = df[["Experience"]]
y = df["Salary"]
```

4.Code (Polynomial Features)

```python
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
```

5.Code (Model Training)

```python
model = LinearRegression()
model.fit(X_poly, y)
```

6.Code (Prediction)

```python
predictions = model.predict(X_poly)
print(predictions)
```

7.code(graph)

```python
plt.scatter(X, y)
plt.plot(X, predictions)
plt.title("Experience vs Salary (Polynomial Regression)")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.grid()
plt.show()
```

8.Code (Custom Prediction)

```python
exp = [[6]]
exp_poly = poly.transform(exp)
pred = model.predict(exp_poly)
print("Predicted salary:", pred[0])
```

## Conclusion

- Polynomial regression captures non-linear relationships  
- It fits the data better than linear regression for curved patterns  
- Useful when data is not in a straight line
